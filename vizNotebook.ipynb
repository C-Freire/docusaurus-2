#### Tal solução foi desenvolvida para substituir um sistema de Ordens de Serviço do setor de Manutenção Industrial que era feito com blocos impressos (com carbono), em que cada O.S. gerava 3 vias de papel

### O back-end, que é uma lista do Sharepoint On Line, foi exportada para o Excel e utilizaremos o pandas para leitura do arquivo e tratamento dos dados

```python
import pandas as pd
import plotly
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import pytz
import plotly.graph_objects as go
%matplotlib inline
```

```python
df = pd.read_excel("dFato.xlsx")
```

```python
pSetor=df['setor'].value_counts()
```

```python
pTipo=df['tipo_os'].value_counts()
```

```python
pSituacao=df['situacao_inicial'].value_counts()
```

```python
pData = df['data'].value_counts()
```

```python
x=pSetor.index
y=pSetor.values
x2=pTipo.index
y2=pTipo.values
x3=pSituacao.index
y3=pSituacao.values
x4=pData.index
y4=pData.values
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(y=y, x=x))
fig.update_layout(title = 'Ordens de Serviço por Setor')
fig.show(figsize=(8,2))
```

```python
fig = px.pie(pSituacao,  values=y3, names=x3, color=x3, color_discrete_map={
    'PARADA POR QUEBRA':'lightcyan',
    'TRABALHANDO':'royalblue',
    'LIBERADO P/ MANUTENÇÃO':'cyan'})
fig.update_layout(title = 'Ordens de Serviço por Tipo') 
fig.show(figsize=(2,2))
```

```python
fig = px.bar(pData, x=x4, y=y4)
fig.update_layout(xaxis_range=['2021-01-01','2021-05-31'],
                  title_text="Manually Set Date Range")
fig.show(figsize=(6,2))
```
